## Modul 7 - Training und Evaluation (Erklaerung)

Ziel: Training starten, Metriken lesen, Overfitting/Underfitting diagnostizieren.

ToDo:

- B: Lies Loss/Accuracy sowie Confusion Matrix aus.
- C: Schreibe 2 Diagnose-Saetze zu den Metriken.

Reflexion nach dem Lauf:
- Was hat sich bei Train/Val-Accuracy veraendert?
- Wo siehst du Hinweise auf Overfitting?
- Welche Klasse ist in der Confusion Matrix auffaellig?

```python
# Modul 7 - Musterloesung (robuster Fallback)

def train_one_epoch(model, loader, criterion, optimizer):
    model.train()
    total_loss, correct, total = 0.0, 0, 0
    for xb, yb in loader:
        xb, yb = xb.to(utils.device), yb.to(utils.device)
        optimizer.zero_grad()
        out = model(xb)
        loss = criterion(out, yb)
        loss.backward()
        optimizer.step()

        total_loss += loss.item() * xb.size(0)
        pred = out.argmax(dim=1)
        correct += (pred == yb).sum().item()
        total += yb.size(0)

    return total_loss / max(total, 1), 100.0 * correct / max(total, 1)

def validate_one_epoch(model, dataloader, criterion):
    model.eval()
    running_loss = 0.0
    correct_predictions = 0
    total_samples = 0
    weighted_conf_correct = 0.0
    weighted_conf_total = 0.0

    with utils.torch.no_grad():
        for batch_images, batch_labels in dataloader:
            batch_images = batch_images.to(utils.device)
            batch_labels = batch_labels.to(utils.device)
            outputs = model(batch_images)
            loss = criterion(outputs, batch_labels)
            running_loss += loss.item() * batch_images.size(0)

            probs = utils.torch.softmax(outputs, dim=1)
            confs, predicted = utils.torch.max(probs, 1)
            correct_predictions += (predicted == batch_labels).sum().item()
            total_samples += batch_images.size(0)
            weighted_conf_correct += confs[(predicted == batch_labels)].sum().item()
            weighted_conf_total += batch_labels.size(0)

    avg_loss = running_loss / max(total_samples, 1)
    accuracy = 100.0 * correct_predictions / max(total_samples, 1)
    conf_weighted_acc = 100.0 * weighted_conf_correct / max(weighted_conf_total, 1)
    return avg_loss, accuracy, conf_weighted_acc

def evaluate(model, loader):
    model.eval()
    all_true, all_pred = [], []
    with utils.torch.no_grad():
        for xb, yb in loader:
            xb = xb.to(utils.device)
            out = model(xb)
            pred = out.argmax(dim=1).cpu().numpy().tolist()
            all_pred.extend(pred)
            all_true.extend(yb.numpy().tolist())
    acc = (utils.np.array(all_true) == utils.np.array(all_pred)).mean() * 100.0
    return acc, all_true, all_pred

def run_training(epochs, lr):
    selected_model_name = utils.model_name_w.value
    selected_training_mode = utils.training_mode_w.value
    selected_optimizer_name = utils.optimizer_name_w.value
    freeze_backbone = selected_training_mode == 'feature_extraction'

    model = utils.build_model(
        num_classes=utils.num_classes,
        model_name=selected_model_name,
        freeze_backbone=freeze_backbone,
    )

    criterion = utils.nn.CrossEntropyLoss()
    trainable_params = filter(lambda p: p.requires_grad, model.parameters())

    if selected_optimizer_name == 'adam':
        optimizer = utils.torch.optim.Adam(trainable_params, lr=lr)
    elif selected_optimizer_name == 'adamw':
        optimizer = utils.torch.optim.AdamW(trainable_params, lr=lr)
    elif selected_optimizer_name == 'sgd':
        optimizer = utils.torch.optim.SGD(trainable_params, lr=lr, momentum=utils.sgd_momentum_w.value)
    else:
        raise ValueError(f'Unsupported optimizer: {selected_optimizer_name}')

    scheduler = utils.torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=epochs, eta_min=lr * 1e-2)
    hist = {'train_loss': [], 'train_acc': [], 'val_loss': [], 'val_acc': []}

    for epoch in range(epochs):
        tr_loss, tr_acc = train_one_epoch(model, utils.train_loader, criterion, optimizer)
        va_loss, va_acc, conf_weighted_acc = validate_one_epoch(model, utils.val_loader, criterion)
        scheduler.step()

        hist['train_loss'].append(tr_loss)
        hist['train_acc'].append(tr_acc)
        hist['val_loss'].append(va_loss)
        hist['val_acc'].append(va_acc)

        current_lr = optimizer.param_groups[0]['lr']
        utils.display_training_status(
            epoch=epoch,
            epochs=epochs,
            selected_model_name=selected_model_name,
            selected_training_mode=selected_training_mode,
            selected_optimizer_name=selected_optimizer_name,
            lr=lr,
            current_lr=current_lr,
            tr_loss=tr_loss,
            tr_acc=tr_acc,
            va_loss=va_loss,
            va_acc=va_acc,
            conf_weighted_acc=conf_weighted_acc,
            hist=hist,
        )

    test_acc, y_true, y_pred = evaluate(model, utils.test_loader)
    print(f'Final Test Accuracy: {test_acc:.2f}%')
    return model, hist, y_true, y_pred

train_btn, output = utils.create_train_button(run_training)
display(train_btn, output)
print('Diagnoseauftrag: Achte auf den Abstand zwischen Train- und Val-Accuracy sowie auf auffaellige Felder in der Confusion Matrix.')
```
